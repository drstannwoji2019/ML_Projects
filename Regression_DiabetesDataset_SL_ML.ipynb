```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel('/content/FemaleDiabetesDataSet.xlsx')

# The first column (index 0) in mnist_test.csv is the label (digit)
# The remaining columns are pixel values, which we will use as features (X)
X = data[['glucose']].values  # Select all columns except the first one
# We need to pass the name of the column as a string to the DataFrame to select it
Y = data[['bmi']].values    # Select the first column as the target variable (label)
#We need to pass the name of the column as a string to the DataFrame to select it

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Fit the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predictions
Y_pred = model.predict(X_test)

# Since we are working with high-dimensional data,
# we cannot directly plot the regression line as in a simple 2D case.
# Instead, we can evaluate the model's performance using metrics like R-squared.
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, Y_pred)
print(f"R-squared: {r2}")

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Fit the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predictions
Y_pred = model.predict(X_test)

# Plotting the regression line
# Instead of plotting against all of X_test, we'll plot against a single feature
# for visualization purposes. Here, we're choosing the first feature (index 0).
# You can change this to any other feature index if desired.
plt.scatter(X_test[:, 0], Y_test, color='blue')  # Plotting against the first feature of X_test
plt.plot(X_test[:, 0], Y_pred, color='red')       # Plotting predictions against the first feature of X_test
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
# Multiple Linear Regression

# Assume the dataset has multiple features like 'X1', 'X2', 'X3'
X = data[['bmi', 'diastolic', 'pregnancies']].values
Y = data['glucose'].values

# Split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Fit the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predictions
Y_pred = model.predict(X_test)

# Evaluating the model
print("Coefficient: ", model.coef_)
print("Intercept: ", model.intercept_)
```
