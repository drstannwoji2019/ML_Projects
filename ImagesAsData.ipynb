```python
!pip install requests

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests # Import the requests library
from io import BytesIO # Import BytesIO to handle binary data in memory

# Load an image from a URL
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/14/Silz_daim1.jpg/500px-Silz_daim1.jpg"

try:
    response = requests.get(image_url)
    response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
    img = Image.open(BytesIO(response.content)) # Open the image from the bytes content
except requests.exceptions.RequestException as e:
    print(f"Error downloading image: {e}")
    # Handle the error, maybe exit or load a default image
    exit()


img = img.convert("RGB")  # Ensure 3 channels
img_array = np.array(img)

print("Image shape:", img_array.shape)
print("Pixel [0,0]:", img_array[0, 0])

plt.imshow(img_array)
plt.axis('off')
plt.title("Image as NumPy Array")
plt.show()
```
