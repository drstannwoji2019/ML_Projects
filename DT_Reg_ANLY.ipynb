```python
# Import libraries
import pandas as pd # Importing the pandas library with the alias 'pd'
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split # Importing train_test_split
import matplotlib.pyplot as plt # Importing matplotlib.pyplot
from sklearn import tree # Importing tree from sklearn

# Creating a hypothetical dataset
data_reg = {
    'Area_sqft': [1500, 2500, 1800, 3000, 1200, 2200, 3500, 2700, 1600, 2900],
    'Bedrooms': [3, 4, 3, 5, 2, 4, 5, 4, 3, 4],
    'Location_Index': [1, 2, 1, 3, 1, 2, 3, 2, 1, 3],  # Categorical, already encoded
    'Price': [250000, 400000, 280000, 500000, 200000, 350000, 550000, 420000, 230000, 480000]  # Target
}

# Convert to DataFrame
df_reg = pd.DataFrame(data_reg)

# Splitting data into features (X) and target (y)
X_reg = df_reg.drop(columns=['Price'])
y_reg = df_reg['Price']

# Splitting the dataset into training and testing sets
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Creating and training the Decision Tree regressor
regressor = DecisionTreeRegressor(criterion="squared_error", max_depth=3, random_state=42)
regressor.fit(X_train_reg, y_train_reg)

# Visualizing the Decision Tree
plt.figure(figsize=(12,6))
tree.plot_tree(regressor, feature_names=X_reg.columns, filled=True)
plt.show()

# Model Evaluation
y_pred = regressor.predict(X_test_reg)
mae = mean_absolute_error(y_test_reg, y_pred)
print(f"Regression Model MAE: {mae:.2f}")
```
