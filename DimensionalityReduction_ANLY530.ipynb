```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Step 1: Load a sample high-dimensional dataset (Digits dataset)
digits = load_digits()
X = digits.data  # 64 features (8x8 images flattened)

# Step 2: Standardize the data (PCA works better with normalized data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA to reduce dimensions (from 64 to 2 for visualization)
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)

# Step 4: Convert to DataFrame for better visualization
df_pca = pd.DataFrame(X_pca, columns=['PCA Component 1', 'PCA Component 2'])
df_pca['Digit Label'] = digits.target  # Add labels for color mapping

# Step 5: Plot the reduced dimensions with digit labels
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=digits.target, cmap='viridis', alpha=0.7, edgecolors='k')
plt.colorbar(label="Digit Label")
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA Dimensionality Reduction (Digits Dataset)')
plt.show()

# Step 6: Display the variance explained by each principal component
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance by Component 1: {explained_variance[0]:.2f}")
print(f"Explained Variance by Component 2: {explained_variance[1]:.2f}")
print(f"Total Variance Explained: {sum(explained_variance):.2f}")
```
