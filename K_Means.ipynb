```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic dataset
# We create a dataset with 300 samples, 3 centers, and 2 features for simplicity.
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=0)

# Step 2: Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 3: Apply K-Means clustering
# Define the K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Step 4: Visualize the clusters
# Plot the data points with colors corresponding to cluster assignments
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=30, label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=30, label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=30, label='Cluster 3')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=100, c='red', label='Centroids', marker='X')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Step 5: Evaluate the clustering using Silhouette Score
silhouette_avg = silhouette_score(X, y_kmeans)
print(f"Silhouette Score for 3 clusters: {silhouette_avg:.2f}")
```
