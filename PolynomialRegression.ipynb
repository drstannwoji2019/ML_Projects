```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Generate some example data
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)  # Random data points between 0 and 5
y = np.sin(X).ravel() + np.random.normal(0, 0.2, X.shape[0])  # Sinusoidal data with noise

# Visualize the original data
plt.scatter(X, y, color='blue', label='Original Data')

# Transform data to include polynomial features (degree 3 in this case)
degree = 3
polynomial_features = PolynomialFeatures(degree=degree)
X_poly = polynomial_features.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict using the polynomial model
y_poly_pred = model.predict(X_poly)

# Calculate and print the performance of the model
mse = mean_squared_error(y, y_poly_pred)
r2 = r2_score(y, y_poly_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")

# Plot the results
plt.plot(X, y_poly_pred, color='red', label=f'Polynomial Regression (degree {degree})')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```
